```python
import requests
import bs4
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pages = []
prices = []

ratings = []
titles = []
urls = []
```

```python
no_of_pages = 1

for i in range(1,no_of_pages + 1):
  url=('http://books.toscrape.com/catalogue/page-{}.html'.format(i))

  pages.append(url)
```

```python
# viewing the page urls

pages
```

```python
for items in pages:
  page = requests.get(items)
  soup = bs4.BeautifulSoup(page.text,'html.parser')
```

```python
soup
```

```python
# making the output look more beautiful
print(soup.prettify())
```

```python
for t in soup.findAll('h3'):
  print(t)
```

```python
# storing all the title in a list elements

for t in soup.findAll('h3'):
  tittles = t.getText()
  titles.append(tittles)
```

```python
titles
```

```python
# find all prices

for p in soup.find_all('p',class_ ='price_color'):
  pricess = p.getText()
  prices.append(pricess)
```

```python
prices
```

```python
# getting all the ratings in the first page
# s stands for the p tag and k stands for the class and rating
for r in soup.find_all('p',class_ = 'star-rating'):
  for s,k in r.attrs.items():
    star = k[1]
    ratings.append(star)



```

```python
ratings
```

```python
# extracting all image urls

div = []
```

```python
div = soup.find_all('div', class_='image_container') # fetching all element with div
for thumbs in div:
  taggs = thumbs.find('img',class_='thumbnail')
  links = 'http://books.toscrape.com/' + str(taggs['src'])
  newlink = links.replace('..','')
  urls.append(newlink)
```

```python
urls
```

#Storing all extracted data into a dictionary


```python
web_data = {
           'Titles':titles,
           'Prices':prices,
           'Ratings':ratings,
           'URL':urls
           }
```

```python
# checking the lenght of each values
```

```python
print(len(titles))
print(len(prices))
print(len(ratings))
print(len(urls))
```

```python
# creating a dataframe using webdata

df = pd.DataFrame(web_data)
```

```python
df
```

```python
# making the index to start from 1 instead of 0

df.index += 1

df
```

```python
# removing the currency infront of each price

df['Prices'] = df['Prices'].str.replace('Â£','')

df
```

```python
df.sort_values(by='Prices',ascending=False)
```

```python
# converting the ratings into intergers

df['Ratings'] = df['Ratings'].replace({
    'One':1,
    'Two':2,
    'Three':3,
    'Four':4,
    'Five':5
})
```

```python
df
```

```python
# checking the data types

df.dtypes
```

```python
# converting the price into float data type

df['Prices'] = df['Prices'].astype(float)
```

```python
df.dtypes
```

```python
# checking correlation between products

df.corr()
```

# Plotting a graph to see the price and rating corelation

```python
plt.figure(figsize=(5,5))
sns.heatmap(df.corr())
```

```python
# plotting a barplot
plt.figure(figsize=(16,8))

sns.barplot(x = df['Prices'] , y = df['Ratings'],data = df )
```

# From the graph above , it can be seen that books with lower prices have more ratings, thereby meaning people tend to buy books which cost less.
