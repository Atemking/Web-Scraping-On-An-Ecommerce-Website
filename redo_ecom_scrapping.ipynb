```python
# import the libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python


url = 'http://books.toscrape.com/catalogue/page-{}.html'.format()
```

```python
# creating variables to store data
pages = []
prices = []
ratings = []
titles = []
urls = []
div = []
```

```python
no_of_pages = 5

for i in range(1,no_of_pages + 1):
  url = ('http://books.toscrape.com/catalogue/page-{}.html'.format(i))

  pages.append(url)
```

```python
# checking if the pages link display properly
pages
```

```python
for page in pages:
  newpage = requests.get(page)
  soup = BeautifulSoup(newpage.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
# getting all the prices

for p in soup.find_all('p', class_ = 'price_color'):
  var = p.getText()
  prices.append(var)
```

```python
# checking if all the prices we stored

prices
```

```python
# getting all the titles
for t in soup.findAll('h3'):
  var = t.getText()
  titles.append(var)
```

```python
# checking if title was stored
titles
```

```python
for r in soup('p', class_ = 'star-rating'):
  for s,k in r.attrs.items():
    star = k[1]
    ratings.append(star)
```

```python
# checking rating

ratings
```

```python
# extracting all the image links



div = soup.find_all('div', class_ = 'image_container')

for thumbs in div:
  taggs = thumbs.find('img', class_ = 'thumbnail')
  Link = 'http://books.toscrape.com/' + str(taggs['src'])
  newlink = Link.replace('..','')
  urls.append(newlink)
```

```python
urls
```

```python
print(len(titles))
print(len(prices))
print(len(ratings))
print(len(urls))
```

# A block which gets all informations for the number of required pages.


```python
# this for loop contains number of pages to be run for each pages stored in list
for page in pages:
  newpage = requests.get(page)
  soup = BeautifulSoup(newpage.text,'html.parser')
  for p in soup.find_all('p', class_ = 'price_color'):    # getting all the prices
    var = p.getText()
    prices.append(var)
  for t in soup.findAll('h3'):                       # getting for all titles
    var = t.getText()
    titles.append(var)
  for r in soup('p', class_ = 'star-rating'):          # getting for all rating
    for s,k in r.attrs.items():
      star = k[1]
      ratings.append(star)

  div = soup.find_all('div', class_ = 'image_container')         # getting all image links

  for thumbs in div:
    taggs = thumbs.find('img', class_ = 'thumbnail')
    Link = 'http://books.toscrape.com/' + str(taggs['src'])
    newlink = Link.replace('..','')
    urls.append(newlink)





```

```python
print(len(titles))
print(len(prices))
print(len(ratings))
print(len(urls))
```

```python
# storing the data in a dictionary

data = {
    'Title':titles,
    'Prices':prices,
    'Rating':ratings,
    'Links':urls
}
```

```python
# creating a dataframe using pandas

df = pd.DataFrame(data)
```
